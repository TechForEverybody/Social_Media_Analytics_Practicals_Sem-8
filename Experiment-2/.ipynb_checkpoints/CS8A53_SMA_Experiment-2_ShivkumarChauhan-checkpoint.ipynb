# Data Collection from Official Himalaya Wellness Site

### 1. Blogs data

```python
def getBlogsData(blog_pages_urls:list)->pandas.DataFrame:
    """funtion to get blogs data

    Args:
        blog_pages_urls (list): list of links of pages where blogs previes are get shown

    Returns:
        pandas.DataFrame: requested blogs data in DataFrame
    """
    blogs_urls=[]
    blog_titles=[]
    blog_posting_timestamp=[]
    blog_contents=[]
    for blog_pages_url in blog_pages_urls:
        request=requests.get(blog_pages_url)
        content=bs4.BeautifulSoup(request.content)
        blog_links=content.find_all("a",class_="dib-post")
        for link in blog_links:
            blogs_urls.append(link['href'])
    for blogs_url in blogs_urls:
        print(blogs_url)
        request=requests.get(blogs_url)
        content=bs4.BeautifulSoup(request.content)
        title=content.find_all("h1",class_="dib-post-title")[0]
        blog_titles.append(title.text.strip())
        posting_timestamp=content.find_all("span",class_="dib-post-date")[0]
        blog_posting_timestamp.append(posting_timestamp.text.strip())
        description=[]
        descriptions=content.find_all(class_="dib-post-content")[0]
        for element in descriptions.find_all():
            if element.name in ["h1","h2","h3"]:
                break
            elif element.name in ["p","strong"]:
                description.append(element.text.strip())
        blog_contents.append(" ".join(description))
    Blog_Data=pandas.DataFrame()
    Blog_Data['name']=blog_titles
    Blog_Data['posting_timestamp']=blog_posting_timestamp
    Blog_Data['content']=blog_contents
    Blog_Data['url']=blogs_urls
    return Blog_Data
```

### 2. Products Data Collection

```python
def getProductsData(initial_url:str,limit=1)->pandas.DataFrame:
    """function to get products data

    Args:
        initial_url (str): initial common url
        limit (int, optional): number of pages to be traversed. Defaults to 1.

    Returns:
        pandas.DataFrame: requested products data in DataFrame
    """
    products_page_urls=[]
    for page_index in range(limit):
        products_page_url=initial_url+str(page_index+1)
        products_page_urls.append(products_page_url)
    products_pages=[]
    for products_page in products_page_urls:
        browser.get(products_page)
        html = browser.page_source
        content = bs4.BeautifulSoup(html, 'lxml')
        product_links=content.find_all("a",class_="snize-view-link")
        for product_link in product_links:
            products_pages.append("https://himalayawellness.in"+product_link["href"])
    return products_pages
    titles=[]
    prices=[]
    descriptions=[]
    urls=[]
    for product_page in products_pages:
        print(product_page)
        urls.append(product_page)
        browser.get(product_page)
        html = browser.page_source
        content = bs4.BeautifulSoup(html, 'lxml')
        try:
            title=content.find_all("h1",class_="pt-title")[0]
            titles.append(title.text.strip())
        except Exception as e:
            print("No title")
            titles.append("")
        try:
            price=content.find_all("span",class_="new-price")[0]
            prices.append(price.text.strip())
        except Exception as e:
            print("No price")
            prices.append("")
        try:
            description =content.find_all("div",class_="pt-layout-text")[0]
            descriptions.append(description.text.strip())
        except Exception as e:
            print("No description")
            descriptions.append("")
    Products_Data=pandas.DataFrame()
    Products_Data['title']=titles
    Products_Data['price']=prices
    Products_Data['description']=descriptions
    Products_Data['url']=urls
    return Products_Data
```

# Data Collection from Twitter

### Search Page tweets Scrapping

```python
def getTweetsFromAnySearchTopic(search_string:str,limit=2)->pandas.DataFrame:
    tweets=[]
    counter=0
    for tweet in twitter.TwitterSearchScraper(search_string).get_items():
        if counter>=limit:
            break
        counter+=1
        tweets.append([
            tweet.date.date(),
            tweet.rawContent,
            tweet.username,
            tweet.likeCount,
            tweet.viewCount,
            tweet.hashtags,
            tweet.url,
            tweet.displayname,
            tweet.user.verified
        ])
    print("Total ",counter," tweets are scraped")
    Tweets_Data=pandas.DataFrame(tweets,columns=[
        "date",
        "content",
        "username",
        "likes",
        "views",
        "hashtags",
        "url",
        "name_of_user",
        "is_verified"
    ])
    return Tweets_Data
```

# Data Collection From Facebook

```python
def getFacebookSearchPostsData(facebook_search_url:str,number_of_scrolls=10)->pandas.DataFrame:
    posters=[]
    likes=[]
    contents=[]
    browser.get(facebook_search_url)
    time.sleep(1)
    for i in range(number_of_scrolls):
        browser.execute_script(f"window.scrollTo(0, {2000*i});")
        time.sleep(1)
    html = browser.page_source
    content = bs4.BeautifulSoup(html, 'lxml')
    container=content.find_all("div",class_="x193iq5w x1xwk8fm")[0]
    for content in container.children:
        like=""
        posted_by=""
        desc=""
        try:
            poster=content.find_all("h3",class_="x1heor9g x1qlqyl8 x1pd3egz x1a2a7pz x1gslohp x1yc453h")
            posted_by=poster[0].text
            description=content.find_all("div",class_="x1iorvi4 x1pi30zi x1swvt13 x1l90r2v")
            desc=description[0].text
            likes_details=content.find_all("span",class_="xt0b8zv x1e558r4")
            like=likes_details[0].text
        except:
            pass
        posters.append(posted_by)
        likes.append(like)
        contents.append(desc)
    Data=pandas.DataFrame()
    Data['posted_by']=posters
    Data['content']=contents
    Data['likes']=likes
    return Data
```

```python
def getYoutubeVideoDataFromChannelLink(channel_url:str)->pandas.DataFrame:
    Data=pandas.DataFrame()
    links=[]
    descriptions=[]
    titles=[]
    views=[]
    upload_dates=[]
    categories=[]
    durations=[]
    playlist = Playlist(channel_url)
    video_links_list=[]
    for video in playlist.videos:
        video_links_list.append(video['link'])
    for video_link in video_links_list:
        video = Video.get(video_link, get_upload_date=True)
        try:
            links.append(video['link'])
        except Exception as e:
            print(e)
            links.append("")
        try:
            titles.append(video['title'])
            print(video['title'])
        except Exception as e:
            print(e)
            titles.append("")
        try:
            durations.append(video['duration']['secondsText'])
        except Exception as e:
            print(e)
            durations.append("")
        try:
            views.append(video['viewCount']['text'])
        except Exception as e:
            print(e)
            views.append("")
        try:
            descriptions.append(video['description'])
        except Exception as e:
            print(e)
            descriptions.append("")
        try:
            upload_dates.append(video['uploadDate'])
        except Exception as e:
            print(e)
            upload_dates.append("")
        try:
            categories.append(video['category'])
        except Exception as e:
            print(e)
            categories.append("")
    Data['link']=links
    Data['title']=titles
    Data['duration']=durations
    Data['views']=views
    Data['description']=descriptions
    Data['upload_date']=upload_dates
    Data['category']=categories
    return Data
```

### Youtube Video Data From Search Result

```python
def getYoutubeVideoFromSearch(search_url:str,number_of_scroll=5)->pandas.DataFrame:
    Data=pandas.DataFrame()
    links=[]
    descriptions=[]
    titles=[]
    views=[]
    upload_dates=[]
    categories=[]
    durations=[]
    video_links_list=[]
    browser.get(search_url)
    time.sleep(1)
    for i in range(number_of_scroll):
        browser.execute_script(f"window.scrollTo(0, {2000*i});")
        time.sleep(1)
    html = browser.page_source
    content = bs4.BeautifulSoup(html, 'lxml')
    video_links=content.find_all("a",class_="yt-simple-endpoint style-scope ytd-video-renderer")
    for video_link in video_links:
        if video_link['href'].find("watch")!=-1:
            video_links_list.append(youtube_link+video_link['href'])
    print("Total",len(video_links_list),"Videos founded")
    for video_link in video_links_list:
        print(video_link)
        video = Video.get(video_link, get_upload_date=True)
        try:
            links.append(video['link'])
        except Exception as e:
            print(e)
            links.append("")
        try:
            titles.append(video['title'])
            print(video['title'])
        except Exception as e:
            print(e)
            titles.append("")
        try:
            durations.append(video['duration']['secondsText'])
        except Exception as e:
            print(e)
            durations.append("")
        try:
            views.append(video['viewCount']['text'])
        except Exception as e:
            print(e)
            views.append("")
        try:
            descriptions.append(video['description'])
        except Exception as e:
            print(e)
            descriptions.append("")
        try:
            upload_dates.append(video['uploadDate'])
        except Exception as e:
            print(e)
            upload_dates.append("")
        try:
            categories.append(video['category'])
        except Exception as e:
            print(e)
            categories.append("")
    Data['link']=links
    Data['title']=titles
    Data['duration']=durations
    Data['views']=views
    Data['description']=descriptions
    Data['upload_date']=upload_dates
    Data['category']=categories
    return Data
```

```python
def getCommentsFromVideoLinks(video_links_list:str):
    Comments_Data=pandas.DataFrame()
    authors=[]
    contents=[]
    video_links=[]
    likes=[]
    for video_link in video_links_list:
        print(video_link)
        video_id=video_link.split("v=")[-1]
        try:
            comments = Comments.get(video_id)
            print("Total",len(comments['result']),"comments found")
            for comment in comments['result']:
                authors.append(comment['author']['name'])
                contents.append(comment['content'])
                video_links.append(video_link)
                likes.append(comment['votes']['simpleText'])
        except Exception as e:
            print(e)
    Comments_Data['video_link']=video_links
    Comments_Data['author']=authors
    Comments_Data['content']=contents
    Comments_Data['likes']=likes
    return Comments_Data
```
